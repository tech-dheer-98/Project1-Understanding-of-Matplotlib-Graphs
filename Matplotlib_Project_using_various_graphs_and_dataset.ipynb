## c) Pie Chart Graph:
A pie chart is a type of graph that represents data in a circular form. It's also known as a circle chart.

```python
# Pie chart Example: 
#Highlight the Sales Trend into Mobile Retails  domains where,
#competiors are (Oneplus,Apple,Samsung,Redmi and Oppo) , show their % contribution in year 2023.
```

```python
plt.figure(figsize=(8,5),dpi=100)
brands=["OnePlus","Apple","Samsung","Redmi","Oppo"]
Popularity=[10000,3500,4500,3200,1080] # This total numbers should be 100%

color=["red","green","darkorange","pink","blue"]
ex=[0,0,0,0,0] # This explodes the particlular brand from a pie.

#Plotting the Pie Chart#
#explode: It is used to remove  a pie from a chart 
#shadow: It is used to give a 3-D effect to a pie chart.
#autopct: It is used to give % contribution in the pie chart
#startangle: It guides the Pie chart from which angle to start.
plt.title("Sales Contribution in Mobile Sector in, 2023",size=12,color="black")
plt.pie(Popularity,labels=brands,explode=ex,colors=color,shadow=True,autopct="%.2f%%",startangle=0) 

# Adjusting legend position
plt.legend(loc=3, bbox_to_anchor=(1, 0.8))
#Use bbox_to_anchor: when you get the operlapping of the chart with Legend

plt.show()
```

###  Visual Insights :
From above Pie Chart we conculde that major Sales is contributed from Oneplus brand contributing to 44.88% of overall Sales,followed by Samsung brand which is contributing 20.20% of overall Sales.

# d) Box Plot Graph: 
A box plot, also known as a box-and-whisker plot, is a graph that summarizes a set of data. It shows the distribution of a data set based on its five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.


### Short Explanation:
 
1) Box Plot contains: Min Fence,Q1@25% ,Median,Q3@75%,Max Fence.
 
2) Q1=(25/100 )*(n+1) ... This is how percentile is calculated ,where n= datavalues,
   Eg : data=[1,2,3,4,5,6,7,8], here n=8

3) Q3=(75/100 )*(n+1) ... This is how percentile is calculated ,where n= datavalues,
   Eg : data=[1,2,3,4,5,6,7,8], here n=8
   
4) Min/Lower Fence= Q1-1.5(IQR)... Formula to calculate LF,IQR(Inter Quartile Range)

5) Max/Upper Fence= Q3+1.5(IQR)... Formula to calculate UF,IQR(Inter Quartile Range)

6) IQR is the range between Q1 and Q3 ,IQR=Q3-Q1


```python
#Dataset:Importing the "Expense" Dataset.
```

```python
# Plot the Box Plot 

plt.boxplot(data["Amount"],labels=["Amount"])


plt.show()
```

### Visual Insights:
Here we can see that one amount 16k is the outlier

# f) Violin Plot Graph :

# g) Stem Plot Graph:
A stem plot, also known as a stem-and-leaf display, is a graphical representation of quantitative data. It is a two-column diagram that organizes data and shows its frequency distribution.


```python
#Dataset:Importing the "Expense" Dataset.
```

```python
plt.stem(data.Amount,bottom=0,linefmt="--",markerfmt="d",orientation="vertical",label="Stem Trend")
plt.plot(data.Amount,linewidth=1,color="green",linestyle="--",label="Graph Trend")
plt.title("Spending Distribution",size=14,color="Magenta")
plt.ylabel("Amounts",size=14,color="magenta")
plt.xlabel("Data Counts",size=14,color="magenta")
plt.legend()

plt.show()
```

# h) Stack Plot Graph:

A stack plot is a linear data plot that's stacked vertically, with each plot stacked on top of another. Stack plots are typically used to generate cumulative plots.

```python
# Example : Suppose we want to Calculate the No of Customers visiting in a resturant on weekly basis 

Days=[1,2,3,4,5,6,7] 

Week1=[20,50,80,40,50,90,100]
Week2=[100,120,50,90,120,80,110]
Week3=[50,80,60,90,64,80,96]
Week4=[50,120,150,190,140,99,124]

label=["Week1","Week2","Week3","Week4"]
color=["darkorange","indigo","green","pink"]

plt.stackplot(Days,Week1,Week2,Week3,Week4,labels=label,colors=color)
plt.title("Customers Visit Weekly Status",size=12,color="black")
plt.xlabel("Days in Week",size=12,color="black")
plt.ylabel("Count of Cutomers Weekwise",size=12,color="black")

plt.legend(loc=2, bbox_to_anchor=(1,1))
plt.show()
```

###  Visuals Insights:
1) Here we can conclude that Week 4 has the highest numbers of cutomers visiting the Resturant.

2) Also we can conclude that Day 7 (Sunday) in a Week seem to be most crowded.

# i) Step Plot Graph :
A step plot, also known as a step line plot or stair plot, is a data visualization style that shows data points as a series of horizontal and vertical steps.

```python
#Example:Here we will take a data of Daily Walk in Km for 5 Consecutive Days.
```

```python
x=[1,2,3,4,5]
y=[5,8,10,15,18]

plt.step(x,y,marker='o',color='green',label='Walks',linestyle='--',where="post")
plt.xlabel("No of Days",color="blue",size=15)
plt.ylabel("Km",color='blue',size=15)
plt.title("Daily Km Walked",color='blue',size=15)
plt.plot(x,y,linestyle='--',label="Consistency",color='red')
plt.xticks([1,2,3,4,5])

plt.legend()

plt.show()
```

### Visual Insights :
From the above Step Graph plotted with Line Graph we can conclude that there is consistent increase in the Trend of Walking.

```python
##Dataset:Importing the "Expense" Dataset.

data.head(2)
```

```python
step_group=data.groupby(["Category"])["Amount"].sum()
step_group
```

```python
# Plotting the Step Graph with Line Graph to check the sum of Expenses Trend.
plt.figure(figsize=(4,4),dpi=100)
plt.step(data.Category.unique(),step_group,marker="o",where='mid',label="Step Graph")
plt.plot(data.Category.unique(),step_group,marker="o",linestyle='--',label="Line Graph")
plt.xlabel("Categories",size=15,color='black')
plt.ylabel("Amount",size=15,color='black')
plt.title("Expenses Trend Analysis",size=15,color='black')
plt.legend()

plt.show()
# Plotting the Pie chart to check the Percentage Contribution

plt.figure(figsize=(4,4),dpi=100)

plt.pie(step_group,labels=data.Category.unique(),autopct="%.2f %%",colors=["lightgreen","orange","magenta","blue","red"])
plt.xlabel("% Contribution of the Expenses",color="black",size=15)

plt.legend(loc=2, bbox_to_anchor=(1,1))

plt.show()
```

### Visual Insights :
1)From the above Step Graph plotted with the Line Graph we can conclude that , the sum of Expenses of Grocery is Higher for a given month.    
2)From the Pie Chart we can also conclude that 65.19% of Expenses is utilized on Grocery and 15.02% is utilized by Essentials for a given month.

# j) Scatter Plot Graph:
A scatter plot is a mathematical diagram that uses Cartesian coordinates to display values for two variables for a set of data. It is also known as a scatterplot, scatter graph, scatter chart, scattergram, or scatter diagram.



```python
# Example : Consider the Highest Sales in a March month from the ordered food.

x=[3,8,10,14,16,18,20,22,25,26,28,29,30,31] # Days on which the Resturant received orders in march month
y=[10,20,30,50,80,40,42,102,150,190,199,250,300,450] #Orders Corresponding to that day

plt.figure(figsize=(10,5),dpi=100) #Size of the Graph
plt.scatter(x,y,label="Sales Trend",marker="d")
plt.title("Sales Trend for March",size=15,color="Red")
plt.xlabel("Date",size=15,color="Red")
plt.ylabel("Sales",size=15,color="Red")
plt.xticks(x)

plt.colorbar() # Displays the Bar of color
plt.legend()
plt.show()
```

# Visual Insights :
From the above Scatter Chart we can conclute that,Sales of the given month is suddenly increasing after 20th of March Month

# k) Subplots:
In Python, subplots are a way to create multiple plots on the same figure. This can be useful for comparing different data sets or for creating more complex visualizations.


```python
##Dataset:Importing the "Expense" Dataset.
data.head(2)
```

```python
# Plotting the Subplots with the expense dataset

#Defining the dimensions of the Graph for all 4 :
plt.figure(figsize=(10,5),dpi=100)


#Plotting the Violin Graph
plt.subplot(2,2,1) # Parameters are rows,columns, chart number
plt.violinplot(data["Amount"])

# Plotting the Common Title to all 
plt.suptitle("Expenses of January,2023",size=20,color="red")


# Plotting the Bar Graph
plt.subplot(2,2,2) # Parameters are rows,columns, chart number
plt.bar(data["Category"],data.Amount)

# Plotting the Histogram 
plt.subplot(2,2,3) # Parameters are rows,columns, chart number
plt.hist(data.Amount,bins=25)


# Plotting the Step Plot 
plt.subplot(2,2,4) # Parameters are rows,columns, chart number
plt.scatter(data["Category"],data.Amount)


# Saving the Graph 
plt.savefig("Expenses_2023.png")
plt.show()



```

# Visual Insights:
1)Violin chart explains that the Expenses is widely spread under 2000Rs.   
2)From the bar Chart we can conclude that bills contribute the highest Expenses.   
3)From the Histogram chart , we get to know there is a outlier in the graph.    
4)From the scatter chart ,we can conclude that the Outlier is present in the Bills Expenses.
